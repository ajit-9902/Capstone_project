## 3.1) Insert records from 42_District_wise_crimes_committed_against_women_2001_2012.csv into a table

```python
db=sqlite3.connect('crime_database.db')
cursor=db.cursor()
cursor.execute("CREATE TABLE Crimes_against_women(State_UT text, District text, Year int, Rape int, Kidnapping_abduction int, Dowry_Deaths int,Assault_on_women_with_intent_to_outrage_her_modesty int,Insult_to_modesty_of_Women int,Cruelty_by_Husband_or_His_Relatives int,Importation_of_Girls int)")
```

```python
with open(r"42_district.csv") as file:
    records=0
    for row in file:
        cursor.execute('INSERT INTO Crimes_against_women VALUES(?,?,?,?,?,?,?,?,?,?)',row.split(','))
        db.commit()
        records +=1
print(records,"record(s) printed")
```

### we have inserted our values into the table Crimes_against_women

## 3.2) Write SQL query to find the highest number of rapes & Kidnappings that happened in which state, District, and year.

```python
result=cursor.execute("SELECT State_UT, District, Year, Rape, Kidnapping_abduction FROM Crimes_against_women WHERE District NOT LIKE '%Total%'ORDER BY (Rape + Kidnapping_abduction) DESC LIMIT 1;")
result.fetchall()
```

#### The data seems to indicate that Murshidabad district in West Bengal has the highest number of reported rape cases (526) and kidnapping cases (441)

## 3.3)Write SQL query to find All the lowest number of rapes & Kidnappings that happened in which state, District, and year

```python
result=cursor.execute("SELECT State_UT, District, Year, Rape, Kidnapping_abduction FROM Crimes_against_women WHERE ( Rape, Kidnapping_abduction) IN ( SELECT MIN(Rape) AS Lowest_Rape,MIN(Kidnapping_abduction) AS Lowest_Kidnapping FROM Crimes_against_women WHERE District NOT LIKE '%Total%')")
result.fetchall()
```

#### The data provided identifies the states and their corresponding districts that had the lowest reported incidents of rape and kidnapping in a given year.
